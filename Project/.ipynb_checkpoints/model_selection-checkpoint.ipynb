# Model selection
---

### Loading data

```python
import pandas as pd
import numpy as np
```

```python
# Training set
X_train = np.load("data/train_set.npy", allow_pickle=True)
y_train = np.load("data/train_labels.npy", allow_pickle=True)

# Testing set
X_test = np.load("data/test_set.npy", allow_pickle=True)
y_test = np.load("data/test_labels.npy", allow_pickle=True)

# Columns names
columns = np.load("data/columns.npy", allow_pickle=True)
```

## Test of a logistic regression classifier

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression().fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("accuracy=", clf.score(X_test, y_test) )
```

## Feature importance

```python
scores = pd.DataFrame({"importance":np.abs(clf.coef_[0,:])}, index=columns).sort_values('importance', ascending=False)
scores
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = sns.barplot(x=scores.importance, y=scores.index, orient = 'h', palette="coolwarm")
plt.title('Feature importance')
plt.show()
```

```python
estimators = pd.DataFrame({"feature": columns, "estimator": clf.coef_[0,:]}).sort_values('estimator', ascending=True)
```

```python
ax = sns.barplot(x="estimator", y="feature", data = estimators, orient = 'h', palette="RdYlGn")
plt.title('Influence on score per feature')
plt.show()
```

## ROC curve

```python
from sklearn.metrics import roc_curve, auc
```

```python
y_score = clf.fit(X_train, y_train).decision_function(X_test)

fpr, tpr, _ = roc_curve(y_test, y_score)

print("AUC =", auc(fpr, tpr))
# ROC Curve
plt.plot(fpr, tpr, color="darkorange")
plt.title("ROC Curve")
plt.xlabel("FPR"); plt.ylabel("TPR")
plt.show()
```
